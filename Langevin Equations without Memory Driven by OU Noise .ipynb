# Langevin Equations without Memory Driven by Ornstein-Uhlenbeck Noise

Consider the Langevin equation: $$\ddot{x}_{t} = \frac{f(x_{t})\eta_{t}}{\tau \epsilon^{1+\mu}} - \frac{g(x_{t})^{2}}{\tau \epsilon^{\mu}} \dot{x}_{t},$$
where $f(x_{t})$, $g(x_{t})$ are some sufficiently smooth functions that might depend on $x_{t}$, $\mu$ and $\tau$ are some constants, $\epsilon$ is a small positive constant and $\eta_{t}$ is the Ornstein-Uhlenbeck process satisfying the SDE:
$$\dot{\eta}_{t} = -\frac{\alpha}{\epsilon^{2}}\eta_{t} + \frac{\sqrt{\lambda}}{\epsilon} \dot{W}_{t}. $$
Here $W_{t}$ denotes the standard Brownian motion, $\alpha$ and $\lambda$ are positive constants and we have set the mass $m := \epsilon^{\mu}$. 

Setting $\mu=2$, we are interested in the limiting solution when $\epsilon \to 0$. This corresponds to the situation when the mass $m$ and the noise correlation time (which is of the order $\epsilon^{2}$) go to zero at the same rate. Analytically, it can be shown that the limiting solution satisfies the following SDE for $x_{t}$:
$$dx_{t} = \frac{\lambda}{2\alpha^{2}} \frac{\left[ff'g^{3}-2f^{2}g^{2}g'-2 \tau \alpha f^2 g' \right]}{g^{5}(\tau \alpha + g^{2})} dt + \sqrt{\frac{\lambda}{\alpha^{2}}} \frac{f}{g^{2}} dW_{t},$$
where it is understood that $f=f(x_{t})$, $g=g(x_{t})$, etc. and $'$ denotes derivative wrt $x_{t}$.

For numerical solutions, we set $\tau=1$, $\alpha=\lambda=2$ and take the initial conditions $x_{0}=v_{0}=\eta_{0}=2$.

```python
%matplotlib inline
import math
import numpy as np
import numpy.random as npr
import matplotlib.pyplot as plt

# Set up grid
T=10.0
N=1000
t,dt=np.linspace(0,T,N+1,retstep=True)

# Get Brownian motion
dW=npr.normal(0.0, np.sqrt(dt),N+1)
dW[0]=0.0
W=np.cumsum(dW)

# Equation parameters and functions
Xzero=2.0
Vzero=2.0
etazero=2.0
tau=1.0
mu=2.0
alpha=2.0
lamda=2.0
epsilon1=1
epsilon2=0.5
def f(X): return X**2
def fd(X): return 2.0*X
def g(X): return fd(X)
def gd(X): return 2.0


# Prelimit Solution
Xmil1=np.empty_like(t)
Vmil1=np.empty_like(t)
etamil1=np.empty_like(t)
Xmil1[0]=Xzero
Vmil1[0]=Vzero
etamil1[0]=etazero

Xmil2=np.empty_like(t)
Vmil2=np.empty_like(t)
etamil2=np.empty_like(t)
Xmil2[0]=Xzero
Vmil2[0]=Vzero
etamil2[0]=etazero

# Limiting Solution
Xlim=np.empty_like(t)
Xlim[0]=Xzero
def alim(X): return lamda*(f(X)*fd(X)*g(X)**3-2*f(X)**2*g(X)**2*gd(X)-2*tau*alpha*f(X)**2*gd(X))/(2*alpha**2*g(X)**5*(tau*alpha+g(X)**2))
def blim(X): return math.sqrt(lamda/alpha**2)*f(X)/g(X)**2
def blimd(X): return math.sqrt(lamda/alpha**2)*(g(X)**2*fd(X)-2*f(X)*g(X)*gd(X))/g(X)**4

for n in range(N):
    Xmil1[n+1]=Xmil1[n]+dt*Vmil1[n]
    Vmil1[n+1]=Vmil1[n]+dt*(f(Xmil1[n])*etamil1[n]/((epsilon1**(1+mu))*tau)-Vmil1[n]*g(Xmil1[n])**2/((epsilon1**mu)*tau))
    etamil1[n+1]=etamil1[n]-dt*alpha*etamil1[n]/epsilon1**2 + dW[n+1]*math.sqrt(lamda)/epsilon1
    
    Xmil2[n+1]=Xmil2[n]+dt*Vmil2[n]
    Vmil2[n+1]=Vmil2[n]+dt*(f(Xmil2[n])*etamil2[n]/((epsilon2**(1+mu))*tau)-Vmil2[n]*g(Xmil2[n])**2/((epsilon2**mu)*tau))
    etamil2[n+1]=etamil2[n]-dt*alpha*etamil2[n]/epsilon2**2 + dW[n+1]*math.sqrt(lamda)/epsilon2
    
    Xlim[n+1]=Xlim[n]+dt*alim(Xlim[n]) + dW[n+1]*blim(Xlim[n]) + 0.5*(blim(Xlim[n])*blimd(Xlim[n])*(dW[n+1]**2-dt))
    
plt.ion()
plt.plot(t,Xmil1,'g-.',label='X(t)')
plt.plot(t,Vmil1,'r-',label='V(t)')
plt.plot(t,Xlim,'y-',label='Xlim(t)')
plt.legend(loc='best')
plt.xlabel('t')
plt.ylabel('Solns')
plt.suptitle('epsilon=1, mu=2', weight='bold',size=16)
plt.show(1)

plt.plot(t,Xmil2,'g-.',label='X(t)')
plt.plot(t,Vmil2,'r-',label='V(t)')
plt.plot(t,Xlim,'y-',label='Xlim(t)')
plt.legend(loc='best')
plt.xlabel('t')
plt.ylabel('Solns')
plt.suptitle('epsilon=0.5, mu=2', weight='bold',size=16)
plt.show(2)
```

Thus, we can see that as $\epsilon$ becomes smaller, the velocity $v_{t} = \dot{x}_{t}$ becomes a 'faster' process than the position $x_{t}$ and the solution $x_{t}$ gets 'closer' to the solution of the SDE as mentioned above.
